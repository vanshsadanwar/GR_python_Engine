```python
import sympy as sp
from sympy import Matrix, symbols
from sympy.tensor.array import MutableDenseNDimArray 
t, r, theta, phi, rho, M, G, c= symbols('t r theta phi rho M G c')

##edit metric below(use above symbols)
g = Matrix([[-(1-(2*G*M)/r),0,0,0],
            [0,1/(1-(2*G*M)/r),0,0],
            [0,0,(r**2),0],
            [0,0,0,(((sp.sin(theta))**2)*r**2)]])
g_inv=g.inv()
dim=4
coords=(t,r,theta,phi)

##Christofell Symbols Gamma(-a,b,c)
#(-a,b,c means a is upper index)
Gamma= MutableDenseNDimArray.zeros(dim,dim,dim)
for i in range (dim):
    for j in range (dim):
        for k in range (dim):
            term=0
            for m in range (dim):
                term=term+(g_inv[i,m]/2)*(-g[j,k].diff(coords[m])+g[j,m].diff(coords[k])+g[m,k].diff(coords[j]))
            Gamma[i,j,k]=sp.simplify(term)
#edit index values below & run
Gamma[1,1,1]
    
```

```python
##Riemann curvature tensor R(-i,j,k,l)
riemann= MutableDenseNDimArray.zeros(dim,dim,dim,dim)
for a in range (dim):
    for b in range (dim):
        for c in range (dim):
            for d in range (dim):
                term1=0
                for e in range (dim):
                    term1=term1+(+Gamma[a,c,e]*Gamma[e,b,d]
                                 -Gamma[a,d,e]*Gamma[e,b,c])
                riemann[a,b,c,d]=sp.simplify(term1+Gamma[a,b,d].diff(coords[c])
                                 -Gamma[a,b,c].diff(coords[d]))
#edit index values below & run
riemann[0,1,0,1]
```

```python
##Ricci Tensor R(i,j)
ricci= MutableDenseNDimArray.zeros(dim,dim)
for b in range (dim):
    for d in range (dim):
        term2=0
        for a in range (dim):
            term2=term2+riemann[a,b,a,d]
        ricci[b,d]=sp.simplify(term2)
#edit index values below & run
ricci[1,1]
```

```python
##Ricci scalar
term3=0
for a in range (dim):
    for b in range (dim):
            term3=term3+g_inv[a,b]*ricci[a,b]
ricci_scalar=term3
#run
ricci_scalar

        
```

```python
##Einstien Tensor G(i,j)
EinsteinTensor=MutableDenseNDimArray.zeros(dim,dim)
for a in range (dim):
    for b in range (dim):
        EinsteinTensor[a,b]=sp.simplify(ricci[a,b]-(ricci_scalar/2)*g[a,b])
#edit index values below & run
EinsteinTensor[1,1]
```

```python
##Kretschman scalar R(a,b,c,d)*R(-a,-b,-c,-d)
#May take longer
Riemann_up= MutableDenseNDimArray.zeros(dim,dim,dim,dim)
for p in range (dim):
    for q in range (dim):
        for r in range (dim):
            for s in range (dim):
                term4=0
                for a in range (dim):
                    for b in range (dim):
                        for c in range (dim):
                            for d in range (dim):
                                term4=term4 + g[a,p]*g_inv[b,q]*g_inv[c,r]*g_inv[d,s]*riemann[a,b,c,d]
                            Riemann_up[p,q,r,s]=sp.simplify(term4)
term5=0
for p in range (dim):
    for q in range (dim):
        for r in range (dim):
            for s in range (dim):
                term5=term5+Riemann_up[p,q,r,s]*riemann[p,q,r,s]
            kret_scalar=term5
kret_scalar
```
